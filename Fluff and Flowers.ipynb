```python
from textblob import TextBlob
import random
from random import choice
import re
```

```python
def adv_adj(filename):
    with open(filename, 'r') as file:
        string_file = file.read().replace('\n', ' ')
    text = TextBlob(string_file)
    count = 0
    adverbs = []
    adjectives = []
    for (word, tag) in text.tags:
        if tag == 'RB':
            cleaned_word = re.sub('[^0-9a-zA-Z]+', '', word)
            if len(cleaned_word) > 3:
                adverbs.append(cleaned_word.lower())
        elif tag == 'JJ':
            cleaned_word = re.sub('[^0-9a-zA-Z]+', '', word)
            if len(cleaned_word) > 3:
                adjectives.append(cleaned_word.lower())
    return (list(set(adverbs)), list(set(adjectives)))
```

```python
def fluff_and_flowers(filename):
    word_list = adv_adj(filename)
    adverbs = word_list[0]
    adjectives = word_list[1]
    with open(filename, encoding='utf-8') as file:
        lines = file.readlines()
    changed_lines = []
    for line in lines:
        blob = TextBlob(line)
        for (word, tag) in blob.tags:
            if tag == 'NN' or tag == 'NNS' or tag == 'NNP' or tag == 'NNPS':
                adjective = random.choice(adjectives)
                if word.isupper():
                    adjective = adjective.upper()
                word_index = line.index(word)
                line = line[:word_index] + adjective + " " + line[word_index:]
            if tag == 'VB' or tag == 'VBD' or tag == 'VBG' or tag == 'VBN' or tag == 'VBP' or tag == 'VBZ':
                adverb = random.choice(adverbs)
                if word.isupper():
                    adverb = adverb.upper()
                word_index = line.index(word)
                line = line[:word_index] + adverb + " " + line[word_index:]
        changed_lines += line
    return ''.join(changed_lines)
```

```python
print(fluff_and_flowers('dante.txt'))
```
